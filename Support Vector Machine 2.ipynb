### Polynomial Functions as Kernel Functions
A kernel function is a way of computing the dot product of two vectors in a higher-dimensional space without explicitly mapping the vectors to that space. This is known as the "kernel trick." Kernel functions enable algorithms to fit non-linear decision boundaries by implicitly working in a higher-dimensional feature space.

A polynomial kernel is a specific type of kernel function. It computes the dot product of two vectors after mapping them into a higher-dimensional space defined by polynomial functions of the original features.

### Effect of Increasing Epsilon on Number of Support Vectors in SVR
In Support Vector Regression (SVR), the epsilon parameter defines a margin of tolerance where no penalty is given to errors. As the value of epsilon increases, the number of support vectors typically decreases because a larger epsilon allows for a wider margin of tolerance. This means the model allows more data points to fall within the margin without being considered support vectors.

### Effect of Kernel Function, C, Epsilon, and Gamma on SVR Performance
Kernel Function: Determines the type of hyperplane used to separate the data. Common choices are linear, polynomial, and radial basis function (RBF).

When to change: Use linear for linearly separable data, polynomial for data with polynomial relationships, and RBF for more complex, non-linear data.
C Parameter: Controls the trade-off between achieving a low error on the training data and minimizing the model complexity.

High C: Less regularization, more focus on fitting the training data (may lead to overfitting).
Low C: More regularization, simpler model (may underfit).
Epsilon Parameter: Specifies the epsilon-tube within which no penalty is associated in the training loss function.

High Epsilon: More tolerance to errors, fewer support vectors (may underfit).
Low Epsilon: Less tolerance to errors, more support vectors (may overfit).
Gamma Parameter (RBF Kernel): Defines how far the influence of a single training example reaches.

High Gamma: Each point’s influence is smaller (more complex model, may overfit).
Low Gamma: Each point’s influence is larger (simpler model, may underfit).

```python
import pandas as pd 
import numpy as np 
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
print(data.DESCR)
```

```python
df = pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df['target'] = data.target
```

```python
df
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.3,random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred  = svc.predict(X_test)
y_pred
```

```python
from sklearn.metrics  import accuracy_score,classification_report,confusion_matrix
print(accuracy_score(y_pred,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
parameter = {'C':[.1,1,5,10],
            'kernel':['linear','poly','sigmoid'],
            'degree':[1,2,3,4],
            'gamma':[.1,.01,.001,.0001]}
```

```python
svc = SVC()
clf = GridSearchCV(svc,param_grid=parameter,cv=5,refit=True,verbose=3)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_params_
model  =  clf.best_estimator_
```

```python
y_pred = model.predict(X_test)
```

```python
print(accuracy_score(y_pred,y_test))
```
